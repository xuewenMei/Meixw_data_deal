```python
import os
import cv2

imfile_path=r'/data_hs/sjwlab/meixuewen/project2/dataset/localduo_data/images'
for I in os.listdir(imfile_path):
    ipath=os.path.join(imfile_path,I)
    minh, minw = 999999999999, 99999999999999
    maxh, maxw = 0, 0
    for imfile in os.listdir(ipath):
        im_path=os.path.join(ipath,imfile)
        img=cv2.imread(im_path,0)
        height, width = img.shape
        f=height*width
        if f>maxh*maxw:
            maxw=width
            maxh=height
        if f<minw*minh:
            minh=height
            minw=width

    print('max: ',maxh,maxw)
    print('min: ',minh,minw)

```
