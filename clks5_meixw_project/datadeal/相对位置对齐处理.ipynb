```python
import os
import cv2
import numpy as np
import concurrent.futures
from threading import Lock

lock = Lock()  # 创建一个锁对象

def crop_image(image, top, left):
    if image is None:
        print("Error: 图像无法加载！")
        return None

    height, width = image.shape[:2]

    new_top = top
    new_left = left
    new_height = height - top  # 假设我们只从顶部裁剪
    new_width = width - left   # 假设我们只从左侧裁剪

    # 裁剪图像
    cropped_image = image[new_top:new_top + new_height, new_left:new_left + new_width]

    return cropped_image

def topleftget(image):
    if image is None:
        print("Error: 图像无法加载！")
        exit(-1)  

    color_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    black_pixel_row = -1
    black_pixel_column = -1

    # 从上到下扫描图像
    for y in range(image.shape[0]):
        row = image[y, :]
        black_pixels_count = np.sum(row == 0)
        ratio_black_pixels = black_pixels_count / row.size
        
        if ratio_black_pixels < 0.98:
            black_pixel_row = y
            color_image = cv2.line(color_image, (0, y), (image.shape[1], y), (0, 0, 255), 1)
            break

    # 从左到右扫描图像
    for x in range(image.shape[1]):
        column = image[:, x]
        black_pixels_count = np.sum(column == 0)
        ratio_black_pixels = black_pixels_count / column.size
        
        if ratio_black_pixels < 0.96:
            black_pixel_column = x
            color_image = cv2.line(color_image, (x, 0), (x, image.shape[0]), (0, 0, 255), 1)
            break

    return black_pixel_row, black_pixel_column

def process_image(imgpath, outpath):
    img = cv2.imread(imgpath, 0)
    top, left = topleftget(img)
    cropim = crop_image(img, top, left)
    
    # 在写入文件前加锁
    with lock:
        print(outpath)
        cv2.imwrite(outpath, cropim)

if __name__ == "__main__":
    imgfile = r'/data_hs/sjwlab/meixuewen/project2/dataset/localduo_data/images_two_many'
    outpath = r'/data_hs/sjwlab/meixuewen/project2/dataset/localduo_data_redeal'
    os.makedirs(outpath, exist_ok=True)

    # 获取所有需要处理的图像文件路径
    tasks = []
    for file in os.listdir(imgfile):
        imgfilepath = os.path.join(imgfile, file)
        out1 = os.path.join(outpath, file)  
        os.makedirs(out1, exist_ok=True)
        for imfile in os.listdir(imgfilepath):
            imgpath = os.path.join(imgfilepath, imfile)
            out2 = os.path.join(out1, imfile)
            tasks.append((imgpath, out2))

    # 使用线程池执行任务
    with concurrent.futures.ThreadPoolExecutor() as executor:
        futures = [executor.submit(process_image, imgpath, outpath) for imgpath, outpath in tasks]
        
        # 等待所有任务完成
        concurrent.futures.wait(futures)
```

```python
###多线程
import numpy as np
import cv2
import os
import threading

# 定义锁
lock = threading.Lock()

def crop_image_from_bottom(img, crop_height):
    if img is None:
        raise ValueError("无法读取图像或路径不正确")

    height, width = img.shape[:2]

    if crop_height <= 0 or crop_height > height:
        raise ValueError("裁剪高度不合理，请选择一个合适的高度")

    cropped_img = img[0:height - crop_height, :]

    return cropped_img

def bottom_black(image):
    if image is None:
        print("Error: 图像无法加载！")
        exit(-1)

    color_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    black_pixel_row = -1
    for y in range(image.shape[0] - 1, -1, -1):
        row = image[y, :]
        black_pixels_count = np.sum(row == 0)
        ratio_black_pixels = black_pixels_count / row.size
        if ratio_black_pixels < 0.9:
            black_pixel_row = y
            break
    t = image.shape[0] - black_pixel_row - 1
    return t

def process_image(impath, out3_im):
    img = cv2.imread(impath, 0)
    bottom = bottom_black(img)
    if bottom == 0:
        cv2.imwrite(out3_im, img)
    else:
        cropim = crop_image_from_bottom(img, bottom)
        cv2.imwrite(out3_im, cropim)

def worker(idfile, imgfile, outpath):
    idfilepath = os.path.join(imgfile, idfile)
    out1_id = os.path.join(outpath, idfile)
    os.makedirs(out1_id, exist_ok=True)
    for viewfile in os.listdir(idfilepath):
        viewpath = os.path.join(idfilepath, viewfile)
        out2_view = os.path.join(out1_id, viewfile)
        os.makedirs(out2_view, exist_ok=True)
        for im in os.listdir(viewpath):
            impath = os.path.join(viewpath, im)
            out3_im = os.path.join(out2_view, im)
            # 使用锁来确保线程安全
            with lock:
                process_image(impath, out3_im)

imgfile = r'/data_hs/sjwlab/meixuewen/project2/dataset/localduo_data_redeal'
outpath = r'/data_hs/sjwlab/meixuewen/project2/dataset/localduo_bottom'
os.makedirs(outpath, exist_ok=True)

# 创建并启动线程
threads = []
for idfile in os.listdir(imgfile):
    thread = threading.Thread(target=worker, args=(idfile, imgfile, outpath))
    threads.append(thread)
    thread.start()

# 等待所有线程完成
for thread in threads:
    thread.join()

print('over!')
```

```python
import cv2
import numpy as np
import os
from concurrent.futures import ThreadPoolExecutor
import threading



def crop_image_from_right(img, crop_width):
    height, width = img.shape[:2]
    if crop_width <= 0 or crop_width > width:
        raise ValueError("裁剪宽度不合理，请选择一个合适的宽度")
    return img[:, 0:width - crop_width]

def crop_image_from_bottom(img, crop_height):
    height, width = img.shape[:2]
    if crop_height <= 0 or crop_height > height:
        raise ValueError("裁剪高度不合理，请选择一个合适的高度")
    return img[0:height - crop_height, :]

def crop_image(image, top, left):
    if image is None:
        print("Error: 图像无法加载！")
        return None
    height, width = image.shape[:2]
    new_top = top
    new_left = left
    new_height = height - top
    new_width = width - left
    return image[new_top:new_top + new_height, new_left:new_left + new_width]

def right_black(image):
    if image is None:
        print("Error: 图像无法加载！")
        exit(-1)
    black_pixel_col = -1
    for x in range(image.shape[1] - 1, -1, -1):
        col = image[:, x]
        black_pixels_count = np.sum(col == 0)
        ratio_black_pixels = black_pixels_count / col.size
        if ratio_black_pixels < 0.9:
            black_pixel_col = x
            break
    return image.shape[1] - black_pixel_col - 1

def bottom_black(image):
    if image is None:
        print("Error: 图像无法加载！")
        exit(-1)
    black_pixel_row = -1
    for y in range(image.shape[0] - 1, -1, -1):
        row = image[y, :]
        black_pixels_count = np.sum(row == 0)
        ratio_black_pixels = black_pixels_count / row.size
        if ratio_black_pixels < 0.9:
            black_pixel_row = y
            break
    return image.shape[0] - black_pixel_row - 1

def topleftget(image):
    if image is None:
        print("Error: 图像无法加载！")
        exit(-1)
    black_pixel_row = -1
    black_pixel_column = -1
    for y in range(image.shape[0]):
        row = image[y, :]
        black_pixels_count = np.sum(row == 0)
        ratio_black_pixels = black_pixels_count / row.size
        if ratio_black_pixels < 0.98:
            black_pixel_row = y
            break
    for x in range(image.shape[1]):
        column = image[:, x]
        black_pixels_count = np.sum(column == 0)
        ratio_black_pixels = black_pixels_count / column.size
        if ratio_black_pixels < 0.96:
            black_pixel_column = x
            break
    return black_pixel_row, black_pixel_column

# 输入和输出路径
imgfile = r'/data_hs/sjwlab/meixuewen/project2/dataset/orglocal/images_detxt_view'
outpath = r'/data_hs/sjwlab/meixuewen/project2/dataset/orglocal/images_detxt_view_cro'
os.makedirs(outpath, exist_ok=True)

# 创建一个线程锁
lock = threading.Lock()

# 定义处理函数
def process_image(impath):
    try:
        # 读取图像
        img = cv2.imread(impath, 0)

        # 从右侧裁剪
        right = right_black(img)
        if right != 0:
            img = crop_image_from_right(img, right)

        # 从底部裁剪
        bottom = bottom_black(img)
        if bottom != 0:
            img = crop_image_from_bottom(img, bottom)

        # 从顶部和左侧裁剪
        top, left = topleftget(img)
        if top != -1 and left != -1:
            img = crop_image(img, top, left)

        # 保存处理后的图像
        with lock:  # 使用锁来同步对输出文件的写入
            out_folder = os.path.dirname(impath).replace(imgfile, outpath)
            os.makedirs(out_folder, exist_ok=True)
            output_path = os.path.join(out_folder, os.path.basename(impath))
            cv2.imwrite(output_path, img)
    except Exception as e:
        print(f"Error processing {impath}: {e}")

# 创建一个线程池
with ThreadPoolExecutor(max_workers=16) as executor:  # 可以根据实际情况调整 max_workers 的大小
    # 遍历文件夹
    for idfile in os.listdir(imgfile):
        idfilepath = os.path.join(imgfile, idfile)
        for viewfile in os.listdir(idfilepath):
            viewpath = os.path.join(idfilepath, viewfile)
            for im in os.listdir(viewpath):
                impath = os.path.join(viewpath, im)
                executor.submit(process_image, impath)

print("所有任务已完成。")
```
